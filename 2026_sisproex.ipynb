```python
# @title ⚙️ Configuração, Leitura e Limpeza dos Dados
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import csv # Importado para ajudar a lidar com textos gigantes
from google.colab import drive
from google.colab import data_table

# Monta o Drive
drive.mount('/content/gdrive', force_remount=True)

# Caminho do arquivo
caminho_padrao = '/content/gdrive/MyDrive'
caminho_especifico = '/CPPS-UNESP/17-Projetos/2026-sisproex'
caminho_completo = caminho_padrao + caminho_especifico
nome_do_arquivo = 'SISPROEX_FRANCA.tsv'
caminho_arquivo = f'{caminho_completo}/{nome_do_arquivo}'

print(f"Tentando ler o arquivo em: {caminho_arquivo}")

# Leitura e Limpeza
try:
    # A SOLUÇÃO ESTÁ AQUI: engine='python' e quoting=csv.QUOTE_NONE
    df = pd.read_csv(
        caminho_arquivo,
        sep='\t',
        dtype=str,
        on_bad_lines='skip',
        engine='python',
        quoting=csv.QUOTE_NONE
    )

    df.columns = df.columns.str.strip().str.lower()

    # Convertendo números
    df['qtde_bolsas_solicitadas'] = pd.to_numeric(df['qtde_bolsas_solicitadas'], errors='coerce').fillna(0)
    df['qtde_bolsas_concedidas_comissao'] = pd.to_numeric(df['qtde_bolsas_concedidas_comissao'], errors='coerce').fillna(0)
    df['ano_base_inicio_realizacao'] = pd.to_numeric(df['ano_base_inicio_realizacao'], errors='coerce').fillna(0).astype(int)

    # Filtrando dados válidos
    df = df.dropna(subset=['coordenador', 'titulo'])
    df = df[(df['ano_base_inicio_realizacao'] >= 2012) & (df['ano_base_inicio_realizacao'] <= 2022)]
    df['Ano_Texto'] = df['ano_base_inicio_realizacao'].astype(str)

    # Preenchendo vazios
    df['grande_area'] = df['grande_area'].fillna('Não Informada')
    df['area_tematica'] = df['area_tematica'].fillna('Não Informada')
    df['departamento_secao'] = df['departamento_secao'].fillna('Não Informado')

    print(f"✅ Base carregada com sucesso! Total de Projetos válidos: {len(df)}")

except Exception as e:
    print(f"❌ ERRO: {e}")
```

# 1. Desempenho e Engajamento Docente 👩‍🏫
Análise de quais professores possuem maior volume de projetos e maior sucesso na captação de bolsas.

```python
# @title Top 15 Coordenadores (Projetos e Bolsas)
# Top Projetos
top_coord = df['coordenador'].value_counts().head(15).reset_index()
top_coord.columns = ['Coordenador', 'Projetos']
top_coord = top_coord.sort_values(by='Projetos', ascending=True)

fig_bar = px.bar(top_coord, x='Projetos', y='Coordenador', orientation='h',
                 title="Top 15 Coordenadores com mais projetos (2012-2022)",
                 color_discrete_sequence=['#1f77b4'], text='Projetos')
fig_bar.update_layout(xaxis_title="Total de Projetos", yaxis_title="", width=800, height=500)
fig_bar.show()

# Top Bolsas
top_bolsas = df.groupby('coordenador')['qtde_bolsas_concedidas_comissao'].sum().reset_index()
top_bolsas = top_bolsas.sort_values(by='qtde_bolsas_concedidas_comissao', ascending=False).head(15)
top_bolsas = top_bolsas.sort_values(by='qtde_bolsas_concedidas_comissao', ascending=True)

fig_bolsas = px.bar(top_bolsas, x='qtde_bolsas_concedidas_comissao', y='coordenador', orientation='h',
                    title="Top 15 Captadores de Bolsas Concedidas",
                    color_discrete_sequence=['#2ca02c'], text='qtde_bolsas_concedidas_comissao')
fig_bolsas.update_layout(xaxis_title="Bolsas Concedidas", yaxis_title="", width=800, height=500)
fig_bolsas.show()
```

```python
# @title 📅 Constância de Projetos por Ano (Mapa de Calor)
heatmap_data = df.groupby(['coordenador', 'Ano_Texto']).size().reset_index(name='Projetos')
heatmap_pivot = heatmap_data.pivot(index='coordenador', columns='Ano_Texto', values='Projetos').fillna(0)

# Filtra professores com pelo menos 3 projetos para o gráfico não ficar ilegível
heatmap_pivot['Total'] = heatmap_pivot.sum(axis=1)
heatmap_pivot = heatmap_pivot[heatmap_pivot['Total'] >= 3].drop(columns=['Total'])

fig_heat = px.imshow(heatmap_pivot, text_auto=True, aspect="auto",
                     color_continuous_scale='Purples',
                     title="Mapa de Calor: Professores com mais de 3 projetos na década",
                     labels=dict(x="Ano", y="", color="Projetos"))
fig_heat.update_xaxes(type='category', side="top")
fig_heat.update_layout(width=1000, height=700)
fig_heat.show()
```

# 2. Histórico e Departamentos 🏢
Como a extensão se comportou ao longo da última década na UNESP Franca.

```python
# @title 📈 Evolução Histórica e Departamentos
projetos_ano = df['Ano_Texto'].value_counts().sort_index().reset_index()
projetos_ano.columns = ['Ano', 'Total de Projetos']

fig_line = px.line(projetos_ano, x='Ano', y='Total de Projetos', markers=True,
                   title="Volume Total de Projetos por Ano",
                   color_discrete_sequence=['#ff7f0e'])
fig_line.update_yaxes(rangemode="tozero")
fig_line.update_layout(width=800, height=400)
fig_line.show()

# Empilhadas por Departamento
dep_ano = df.groupby(['Ano_Texto', 'departamento_secao']).size().reset_index(name='Projetos')
fig_stack = px.bar(dep_ano, x='Ano_Texto', y='Projetos', color='departamento_secao',
                   title="Projetos por Departamento ao longo do tempo")
fig_stack.update_layout(legend=dict(orientation="h", yanchor="bottom", y=-0.5, xanchor="center", x=0.5), width=900, height=600)
fig_stack.show()
```

# 3. Áreas Temáticas e Financiamento 🎯
Para onde a extensão está olhando (temas) e como está a relação entre pedidos de bolsas e concessões reais.

```python
# @title 🌳 Distribuição de Áreas e Comparativo de Bolsas
df_tree = df.dropna(subset=['grande_area', 'area_tematica'])
fig_tree = px.treemap(df_tree, path=[px.Constant("UNESP Franca"), 'grande_area', 'area_tematica'],
                      title="Proporção por Grande Área e Área Temática",
                      color='grande_area', color_discrete_sequence=px.colors.qualitative.Pastel)
fig_tree.update_traces(root_color="lightgrey")
fig_tree.update_layout(margin=dict(t=50, l=25, r=25, b=25), width=900, height=500)
fig_tree.show()

# Bolsas Solicitadas x Concedidas
bolsas_ano = df.groupby('Ano_Texto')[['qtde_bolsas_solicitadas', 'qtde_bolsas_concedidas_comissao']].sum().reset_index()
fig_bolsas_ano = go.Figure()
fig_bolsas_ano.add_trace(go.Bar(x=bolsas_ano['Ano_Texto'], y=bolsas_ano['qtde_bolsas_solicitadas'], name='Solicitadas', marker_color='indianred'))
fig_bolsas_ano.add_trace(go.Bar(x=bolsas_ano['Ano_Texto'], y=bolsas_ano['qtde_bolsas_concedidas_comissao'], name='Concedidas', marker_color='lightsalmon'))

fig_bolsas_ano.update_layout(title='Bolsas Solicitadas vs Concedidas por Ano', barmode='group', xaxis_title="Ano", yaxis_title="Quantidade de Bolsas", width=900, height=500)
fig_bolsas_ano.show()
```
